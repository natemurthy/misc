```python
# following along from https://twitter.com/pyquantnews/status/1608452332403949571

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import yfinance as yf

from sklearn.decomposition import PCA
```

```python
symbols = ['IBM', 'MSFT', 'META', 'INTC', 'NEM', 'AU', 'AEM', 'GFI']
data = yf.download(symbols, start='2020-01-01', end='2021-11-30')
data.head()
```

```python
portfolio_returns = data['Adj Close'].pct_change().dropna()

pca = PCA(n_components=3)
pca.fit(portfolio_returns)
pct = pca.explained_variance_ratio_
pca_components = pca.components_
```
